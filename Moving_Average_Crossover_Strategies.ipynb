```python
from pandas_datareader import data 
from datetime import datetime
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
starttime = datetime(2018,3,1)
endtime = datetime(2019,3,10)
```

```python
df = data.DataReader(name="GOOG", data_source="yahoo", start=starttime, end=endtime)
```

```python
df_close =pd.DataFrame(df.Close)
```

```python
df_close['Ma9']=df_close.Close.rolling(9).mean()
df_close['Ma21']=df_close.Close.rolling(21).mean()
```

```python
# for simple moving average 
plt.figure(figsize=(15,10))
plt.grid(True)
plt.plot(df_close['Close'],label='google')
plt.plot(df_close['Ma9'],label='google 9day SMA')
plt.plot(df_close['Ma21'],label='google 21day SMA')
plt.legend(loc=2)
plt.show()
```

```python
#for exponential moving average
df_close['exp1'] = df_close.Close.ewm(span=9, adjust=False).mean()
df_close['exp2'] = df_close.Close.ewm(span=21, adjust=False).mean()
```

```python
plt.figure(figsize=(15,10))
plt.grid(True)
plt.plot(df_close['Close'],label='google')
plt.plot(df_close['exp1'],label='google 9Day EMA')
plt.plot(df_close['exp2'], label='google 21Day EMA')
plt.legend(loc=2)
plt.show()
```
